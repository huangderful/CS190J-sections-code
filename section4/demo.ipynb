```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Define your edge list
edge_list = [('A', 'C'), ('C', 'B'), ('C', 'D'),
('C', 'E'), ('D', 'E'), ('D', 'F'), ('F', 'G'),
('F', 'H')]

# Add edges to the graph
G.add_edges_from(edge_list)

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', font_weight='bold', node_size=1500)
plt.show()
# Print the graph's nodes and edges
print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

```python
centrality_metric = nx.degree_centrality(G)

# Print degree centrality of each node
for node, centrality in centrality_metric.items():
    print(f"Node {node}: Degree Centrality = {centrality}")
```

```python
centrality_metric = nx.closeness_centrality(G)

# Print degree centrality of each node
for node, centrality in centrality_metric.items():
    print(f"Node {node}: Closeness Centrality = {centrality}")
```
